```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from pretty_confusion_matrix import pp_matrix
pd.set_option('display.max_rows', 600)
```

```python
df_lora_pred = pd.read_csv("./mobileODT100%_lora_large_clip_predictions.csv")
df_google_pred = pd.read_csv("./mobileODT100%_googlenet_predictions.csv")
df_densenet_pred = pd.read_csv("./mobileODT100%_densenet121_predictions.csv")
df_resnext50_pred = pd.read_csv("./mobileODT100%_resnext50_predictions.csv")
df_resnext101_pred = pd.read_csv("./mobileODT100%_resnext101_predictions.csv")
df_resnet50_pred = pd.read_csv("./mobileODT100%_resnet50_predictions.csv")
df_resnet101_pred = pd.read_csv("./mobileODT100%_resnet101_predictions.csv")
true_label = df_lora_pred["True Label"]
df_lora_pred
```

```python
print(classification_report(true_label, df_lora_pred["Predicted Label"], target_names=["Type 1", "Type 2", "Type 3"]))
confusion_matrix(true_label, df_lora_pred["Predicted Label"])
```

```python
w_lora = 0.5
w_resnext50 = 0.1
w_resnext101 = 0.1
w_google = 0.15
w_dense = 0.15

df_ensemble = df_lora_pred.copy()
df_ensemble = df_ensemble.drop(["Predicted Label", "Type_1_score", "Type_2_score", "Type_3_score"], axis=1)
df_ensemble["lora"] = df_lora_pred["Predicted Label"]
df_ensemble["google"] = df_google_pred["Predicted Label"]
df_ensemble["densenet"] = df_densenet_pred["Predicted Label"]
df_ensemble["resnext50"] = df_resnext50_pred["Predicted Label"]
df_ensemble["resnext101"] = df_resnext101_pred["Predicted Label"]
df_ensemble["resnet50"] = df_resnet50_pred["Predicted Label"]
df_ensemble["resnet101"] = df_resnet101_pred["Predicted Label"]

df_ensemble["equal_ensemble"] = df_ensemble[["lora", "google", "densenet", "resnext50", "resnext101", "resnet50", "resnet101"]].mode(axis=1)[0]
df_ensemble

print(classification_report(df_ensemble["True Label"], df_ensemble["equal_ensemble"], target_names=["Type 1", "Type 2", "Type 3"]))
confusion_matrix(df_ensemble["True Label"], df_ensemble["equal_ensemble"])
```
